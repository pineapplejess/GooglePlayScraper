```python
import pandas as pd
import numpy as np
from google_play_scraper import app, Sort, reviews_all
import plotly.express as px
```

```python
gpscraper_project = reviews_all('com.pdfconverter.imagetopdf.pdfreader', sleep_milliseconds=0, lang='en', country='ES', sort=Sort.NEWEST)
```

```python
df = pd.json_normalize(gpscraper_project)
```

```python
from transformers import pipeline
sentiment_analysis = pipeline('sentiment-analysis', model="siebert/sentiment-roberta-large-english")
```

```python
print(sentiment_analysis('It has too many ads'))
```

```python
df.dtypes
```

```python
df['content'] = df['content'].astype('str') #convert to string
```

```python
df['result']=df['content'].apply(lambda x: sentiment_analysis(x) )
```

```python
df.head()
```

```python
df.shape
```

```python
df['sentiment'] = df['result'].apply(lambda x:(x[0]["label"]))
df['score'] = df['result'].apply(lambda x:(x[0]["score"]))
#separate into 2 columns
```

```python
df['sentiment'].value_counts()
```

```python
df['sentiment'].value_counts(normalize=True)
```

```python
fig = px.histogram(df, x='sentiment', color ='sentiment', text_auto=True) #plotlib
fig.show()
```
